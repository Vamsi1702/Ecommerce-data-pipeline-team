```python
df_orders = spark.table("`e-commerce_sales_catalog`.bronze.bronze_orders")
```

```python
df_orders.printSchema()
print("Initial Count:", df_orders.count())
```

```python
df_orders = df_orders.dropDuplicates(["order_id"])
```

```python
from pyspark.sql.functions import col, coalesce, lit

df_orders = df_orders.withColumn(
    "order_status",
    coalesce(col("order_status"), lit("unknown"))
)
```

```python
from pyspark.sql.functions import to_timestamp

df_orders = df_orders.withColumn(
    "order_purchase_timestamp",
    to_timestamp("order_purchase_timestamp")
).withColumn(
    "order_approved_at",
    to_timestamp("order_approved_at")
).withColumn(
    "order_delivered_carrier_date",
    to_timestamp("order_delivered_carrier_date")
).withColumn(
    "order_delivered_customer_date",
    to_timestamp("order_delivered_customer_date")
).withColumn(
    "order_estimated_delivery_date",
    to_timestamp("order_estimated_delivery_date")
)
```

```python
df_orders.display(5)
df_orders.printSchema()
print("Final Count:", df_orders.count())
```

```python
from pyspark.sql.functions import initcap

df_orders = df_orders.withColumn(
    "order_status",
    initcap(col("order_status"))
)
```

```python
df_orders.filter(col("order_purchase_timestamp").isNull()).show()
```

```python
df_orders.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("`e-commerce_sales_catalog`.silver.silver_orders")
```
