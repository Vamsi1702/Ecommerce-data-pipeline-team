```python
df_customers = spark.table("`e-commerce_sales_catalog`.bronze.bronze_customers")
```

```python
df_customers.printSchema()
print("Initial Count:", df_customers.count())
```

```python
df_customers = df_customers.dropDuplicates(["customer_id"])
```

```python
from pyspark.sql.functions import col, coalesce, lit

df_customers = df_customers.withColumn(
    "customer_city",
    coalesce(col("customer_city"), lit("unknown"))
)

df_customers = df_customers.withColumn(
    "customer_state",
    coalesce(col("customer_state"), lit("unknown"))
)
```

```python
from pyspark.sql.functions import initcap

df_customers = df_customers.withColumn(
    "customer_city",
    initcap(col("customer_city"))
)
```

```python
df_customers.display()
print("Final Count:", df_customers.count())
```

```python
df_customers.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("`e-commerce_sales_catalog`.silver.silver_customers")
```
