```python
df_customers = spark.table("`e-commerce_sales_catalog`.silver.silver_customers")
df_products = spark.table("`e-commerce_sales_catalog`.silver.silver_products")
df_orders = spark.table("`e-commerce_sales_catalog`.silver.silver_orders")
df_order_items = spark.table("`e-commerce_sales_catalog`.silver.silver_order_items")
```

```python
df_sales = df_order_items
```

```python
df_sales = df_sales.join(
    df_orders,
    on="order_id",
    how="inner"
)
```

```python
df_sales = df_sales.join(
    df_customers,
    on="customer_id",
    how="inner"
)
```

```python
df_sales = df_sales.join(
    df_products,
    on="product_id",
    how="inner"
)
```

```python
df_sales = df_sales.select(
    "order_id",
    "order_item_id",
    "customer_id",
    "customer_city",
    "customer_state",
    "product_id",
    "product_category_name",
    "price",
    "freight_value",
    "order_status",
    "order_purchase_timestamp",
    "order_delivered_customer_date"
)
```

```python
from pyspark.sql.functions import col, round

df_sales = df_sales.withColumn(
    "total_amount",
    round(col("price") + col("freight_value"), 2)
)
```

```python
df_sales.display()
print("Final Count:", df_sales.count())
df_sales.printSchema()
```

```python
df_sales.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("`e-commerce_sales_catalog`.silver.silver_sales_detail")
```
